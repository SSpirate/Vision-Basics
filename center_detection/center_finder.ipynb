**Locating center of the circle**

The center of the circle can be defined by locating the end points of the diameter. The following algorithm locates the end points of a horizontal diameter.
The process involves the following steps : 

- Converting the image into a 2 dimensional array.
- Identifying the pixels which define the circumference of the circle.
- Locating farthest points on the diameter by traversing row wise.
- Applying mid point formula for locating the center.


```python
#Converting the image into a 2d aray by taking the mean across depth axis
test_image = np.mean(image, axis=2)
```

```python
#Displaying the generated image.
plt.imshow(test_image)
```

```python
#Locating pixels with darkest and brightest value.
max_value = np.max(test_image)
min_value = np.min(test_image)
(max_value, min_value)
```

Here, the image contains only 2 pixel values(black and white).

The black pixels are represented by the minimum value in the array and hence they are located using the same logic.

```python
#Locating the circumference
X, y = list(), list()
for i in range(test_image.shape[0]):
    for j in range(test_image.shape[1]):
        if test_image[i][j] == min_value:
            X.append(i)
            y.append(j)
```

```python
len(X)
```

Here, we find that 6947 pixels were found on the circumference of the circle

```python
#representing the circumference
plt.scatter(X, y)
plt.show()
```

```python
#Algorithm for locating farthest points(horizontal)
distance = -1
y1_h = -1
y2_h = -1
x1_h = -1
for i in range(test_image.shape[0]):
    pixel_list = list()
    for j in range(test_image.shape[1]):
        if test_image[i][j] == min_value:
            pixel_list.append(j)
    if len(pixel_list)!=0 and abs(pixel_list[-1] - pixel_list[0]) > distance:
        y1_h = min(pixel_list)
        y2_h = max(pixel_list)
        x1_h = i
        distance = y2_h - y1_h
```

```python
(y1_h, y2_h, x1_h, distance)
```

Here, the distance parameter represents the pixel length of the diameter. The co-ordinates y1_h and y2_h represent the end points of the diameter.

```python
#Storing the co-ordinates in lists for plotting,
h_plotting_X = [x1_h, x1_h]
h_plotting_y = [y1_h, y2_h]
```

```python
#Plotting the located diameter
plt.scatter(X, y)
plt.plot(plotting_X, plotting_y, color="red")
plt.show()
```

The mid point formula in co ordinate geometry provides the location of a line segment defined by 2 points in a plane. Hence, the co-ordinates of the mid point of the diameter i.e. the center of the circle is given as : 

y = (y1_h + y2_h)/2


```python
#Center
y_c = (y1_h + y2_h)/2
x_c = x1_h
```
